# Dictionary Exercise: How many

```python
def how_many(aDict):
    '''
    aDict: A dictionary, where all the values are lists.

    returns: int, how many values are in the dictionary.
    '''
    # Your Code Here
    howMany = 0
    dictLists = aDict.values() #dictLists is a list of lists
    
    for listElem in dictLists:
        howMany += len(listElem)
        
    return howMany
    
```

```python
animals = { 'a': ['aardvark'], 'b': ['baboon'], 'c': ['coati']}

animals['d'] = ['donkey']
animals['d'].append('dog')
animals['d'].append('dingo')

how_many(animals)
```

```python
def biggest(aDict):
    '''
    aDict: A dictionary, where all the values are lists.

    returns: The key with the largest number of values associated with it
    '''
    biggestKey = ''
    
    if len(aDict) == 0:
        return None
        
    else:
        biggestKey = max(aDict)
        return aDict[biggestKey]
```

```python
animals = { 'a': ['aardvark'], 'b': ['baboon'], 'c': ['coati']}

animals['d'] = ['donkey']
animals['d'].append('dog')
animals['d'].append('dingo')

biggest(animals)
```

# Fibonacci with Dictionary

```python
def fib_efficient(num , aDictMemo = {1: 1, 2: 2}):
    """
    Takes in: an int as 'num' 
    
    Has default setting for aDictMemo. User need not specify it.
    
    Retruns: Fibonacci number corresponding to 'num'
    """
    
#     Keeping track of function calls
    global numFibCalls 
    #This is defined outside of function scope in global scope
    global fullDict
    
    fullDict = aDictMemo
    numFibCalls += 1
    
    #num here refers to the key and NOT to the values
    if num in aDictMemo:
        return aDictMemo[num]
    
#     aDictMemo will be updated based on the return statements during recursion
    else:
        ans = fib_efficient( num-1, aDictMemo ) + fib_efficient( num-2, aDictMemo )
        aDictMemo[num] = ans
#         print(aDictMemo.values())
        return ans
```

```python
def fib(x):
    """
    Assumes x as int, x>0
    
    Returns Fibonacci number
    """
    
    #Keeping track of num function calls
    global numFibCalls
    numFibCalls += 1
    
    if x==0 or x==1:
        return 1
    
    else:
        return fib(x-1) + fib(x-2)
```
